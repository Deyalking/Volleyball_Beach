```python

!pip install -q mediapipe opencv-python

import cv2
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow

mp_pose = mp.solutions.pose
pose = mp_pose.Pose(static_image_mode=False, model_complexity=2, enable_segmentation=False, min_detection_confidence=0.5)

cap = cv2.VideoCapture(video_name)
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter("3D_BEACH_DEMO.mp4", cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose.process(rgb_frame)

    if results.pose_landmarks:

        mp.solutions.drawing_utils.draw_landmarks(
            frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp.solutions.drawing_utils.DrawingSpec(color=(0,255,0), thickness=2),
            connection_drawing_spec=mp.solutions.drawing_utils.DrawingSpec(color=(255,0,0), thickness=2))

        landmarks = results.pose_landmarks.landmark
        knee_angle = np.degrees(np.arccos(
            np.dot(
                np.array([landmarks[23].x - landmarks[25].x, landmarks[23].y - landmarks[25].y]),
                np.array([landmarks[27].x - landmarks[25].x, landmarks[27].y - landmarks[25].y])
            )
        ))
        cv2.putText(frame, f"Knee: {knee_angle:.1f}°", (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (0,255,0), 5)

    out.write(frame)

cap.release()
out.release()
print("3D_BEACH_DEMO.mp4")
```
